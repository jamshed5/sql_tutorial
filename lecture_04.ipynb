```python
# inner join
%sql SELECT employees.emp_id, employees.manager_id, employees.f_name, employees.l_name, employees.address, departments.dep_name FROM employees INNER JOIN departments ON employees.manager_id=departments.manager_id; 
```

```python
# left join
%sql SELECT employees.emp_id, employees.manager_id, employees.f_name, employees.l_name, employees.address, departments.dep_name FROM employees LEFT JOIN departments ON employees.manager_id=departments.manager_id; 
```

```python
# right join 
%sql SELECT employees.emp_id, employees.manager_id, employees.f_name, employees.l_name, employees.address, departments.dep_name FROM employees RIGHT JOIN departments ON employees.manager_id=departments.manager_id; 
```

```python
# full join 
%sql SELECT employees.emp_id, employees.manager_id, employees.f_name, employees.l_name, employees.address, departments.dep_name FROM employees FULL JOIN departments ON employees.manager_id=departments.manager_id; 
```
