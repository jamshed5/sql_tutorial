# Order By

```python
import sqlalchemy as sa
engine = sa.create_engine('postgresql://postgres:pakistan123@localhost:5432/myinfo')
%load_ext sql
%sql $engine.url
```

```python
# Order by name
%sql SELECT * FROM employees ORDER BY f_name;
```

```python
# Order by name -- decending order
%sql SELECT * FROM employees ORDER BY f_name DESC;
```

```python
# order by using multiple columns f_name, address
%sql SELECT * FROM employees ORDER BY f_name,address;
```

```python
# order by using multiple columns f_name(ascending), emp_id(decending)
%sql SELECT * FROM employees ORDER BY f_name ASC, emp_id DESC;
```
